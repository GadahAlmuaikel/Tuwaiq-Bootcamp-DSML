### Question 1: Plotting a Line Graph
**Question:**
Create a line graph for the function $y = 3x + 2$ using NumPy to generate the `x` values from 0 to 10.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,50)
y = 3*x + 2

plt.plot(x,y)
plt.title('line graph for y = 3x + 2')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

### Question 2: Plotting Multiple Functions
**Question:**
Plot the sine and cosine functions on the same graph for $x$ values from 0 to $2\pi$ using NumPy.

Hint

use :x = np.linspace(0, 2 * np.pi, 100)

```python
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)


plt.plot(x,y1,label='sine function')
plt.plot(x,y2,label='cosine function')
plt.title('sine and cosine functions')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

### Question 3: Creating a Histogram
**Question:**
Generate 1000 random numbers from a standard normal distribution using NumPy and plot their histogram using Matplotlib.

```python
data = np.random.randn(1000)
plt.hist(data,bins=50,color='red',edgecolor='black')
plt.title('histogram')
plt.xlabel('normal distribution')
plt.ylabel('frequencey')
plt.show()
```

### Question 4: Customizing Plot Appearance
**Question:**
Plot the function $y = x^2$ and customize the plot with a red dashed line, circle markers, and a grid.

Hint use x = np.linspace(-10, 10, 100)


```python
x = np.linspace(-10, 10, 100)
y=x**2
plt.plot(x,y,color='red',linestyle='--',marker='o')
plt.grid()
plt.title('plot for function  y=x^2 ')
plt.show()
```

### Question 5: Subplots in Matplotlib
**Question:**
Create a 1x2 subplot layout with the first subplot displaying $y = \sin(x)$ and the second subplot displaying $y = \cos(x)$ for $x$ values from 0 to $2\pi$.

```python

x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title('y=sin(x)')
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('y=cos(x)')
plt.subplots_adjust(wspace=0.5)
plt.show()
```

### Question 6: Scatter Plot
**Question:**
Create a scatter plot of 100 random points with x and y values between 0 and 1. Use different colors and sizes for the points.

```python
x = np.random.rand(100)
y= np.random.rand(100)
colors = np.random.rand(100,3)
sizes = np.random.rand(100)*10
plt.scatter(x,y,color=colors,s=sizes)
plt.title('scatter plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
